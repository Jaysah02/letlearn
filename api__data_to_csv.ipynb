```python
import pandas as pd
import requests
```

```python
df1=pd.DataFrame()
```

```python
for i in range(1,487):
  response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=a86e4daf24704ebc1142219998bd676b&language=en-US&page={}'.format(i))
  temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df1=pd.concat([df1,temp_df],ignore_index=True)
```

```python
df1
```

```python
df1.to_csv('movies.csv')
```
